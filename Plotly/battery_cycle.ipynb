```python
import plotly.express as px
import pandas as pd
import plotly.graph_objects as go
```

Read data from metadata file

```python
df=pd.read_csv(r'C:\Users\shibi\Desktop\Python\Learn\Plotly\cleaned_dataset\metadata.csv')
df.head()
```

```python
df.info()
```

Taking the impedance data from rows(type=impedance) with non-null columns of Re and Rct.


```python
imp_data=df[(df['type']=='impedance') & df[['Re','Rct']].notna().all(axis=1)]
```

```python
imp_data=imp_data.sort_values(by='start_time')
```

Elctrolyte Resistance(Re) vs Time Plot


```python
fig_Re=px.line(
    imp_data,
    x='start_time',
    y='Re',
    labels={'start_time':'Time', 'Re':'Elctrolyte Resistance (Ohms)'},
    title='Elctrolyte Resistance(Re) vs Time '
)
fig_Re.show()
```

Charge transfer resistance(Rct) vs Time Plot:

```python
fig_Rct=px.line(
    imp_data,
    x='start_time',
    y='Rct',
    labels={'start_time':'Time ', 'Rct':'Elctrolyte Resistance (Ohms)'},
    title='Charge transfer resistance(Rct) vs Time '
)
fig_Rct.show()
```

Battery Impedance vs Time Plot:

```python
imp_data['Battery_impedance']=imp_data['Rct']+imp_data['Re']

fig_BatteryImp=px.line(
    imp_data,
    x='start_time',
    y='Battery_impedance',
    labels={'start_time':'Time ', 'Battery_impedance':'Battery Impedance (Ohms)'},
    title='Battery Impedance vs Time '
)
fig_BatteryImp.show()
```

Combined Plot:

```python
fig_combined=go.Figure()

fig_combined.add_trace(go.Scatter(
    x=imp_data['start_time'], y=imp_data['Re'],
    mode='lines', name='Electrolyte Resistance (Re)'
))

fig_combined.add_trace(go.Scatter(
    x=imp_data['start_time'], y=imp_data['Rct'],
    mode='lines', name='Charge Transfer Resistance (Rct)'
))

fig_combined.add_trace(go.Scatter(
    x=imp_data['start_time'], y=imp_data['Battery_impedance'],
    mode='lines', name='Battery Impedance (Re + Rct)'
))
```
